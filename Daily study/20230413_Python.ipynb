### 머신러닝 클래스, 하이퍼파라미터 튜닝 시, 입력값이 딕셔너라를 기본값으로 받음

```python
#머신러닝 클래스, 하이퍼파라미터 튜닝 시, 입력값이 딕셔너라를 기본값으로 받음

def temp(**kwargs):
  for key, value in kwargs.items():
    print(key, value)

temp(name = "SU", age = 30, city = "SUWON")
```

```python
## 다중조건식
cols = ['CRIM','ZN','target']

#CRIM 1보다 작고, target이 24 이상인 값만 조회

result = bos.loc[(bos['CRIM'] < 1) &
                 (bos['target'] >= 24), cols]
print(result)

#CRIM 1보다 작거나, target이 24 이상인 값만 조회

result = bos.loc[(bos['CRIM'] < 1) |
                 (bos['target'] >= 24), cols]
print(result)
```

## shift()

```python
temp_df = sales[['datesold', 'price']].copy()
temp_df['shifted_v1'] = temp_df['price'].shift(1, fill_value = 0).astype(int)
temp_df['shifted_v2'] = temp_df['price'].shift(2, fill_value = 0).astype(int)

temp_df['차이값']= temp_df['price'] - temp_df['shifted_v1']
temp_df.head()
```

```python
# 0000년 00월 00일
temp_df['한글날짜'] = temp_df['datesold'].dt.strftime('%Y년 %m월 %d일')
temp_df.head()
```
